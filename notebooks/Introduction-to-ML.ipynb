# Basic 3D Operations

```python
import numpy as np
from copy import copy
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
```

```python
def create_object(vertices, facecolor=[1,1,1], edgecolor=[0,0,0]):
    obj = Poly3DCollection(vertices, linewidths=1, alpha=0.2)
    obj.set_facecolor(facecolor)
    obj.set_edgecolor(edgecolor)
    return obj
```

```python
def get_plots(objs):
    plt.ion()
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    for obj in objs:
        print(obj)
        ax.add_collection3d(copy(obj))
    s = [-2,-2,-2,2,2,2]
    ax.auto_scale_xyz(s,s,s)
    return ax, fig
```

```python
objs = []
```

### Irregular tetrahedron

```python
corners = [[-1,0.5,0.5], [+1,0.5,0.5], [0,-0.5,0.5], [0,0.5,-0.5]]
```

```python
face1 = [ corners[0], corners[1], corners[2] ] 
face2 = [ corners[0], corners[1], corners[3] ] 
face3 = [ corners[0], corners[2], corners[3] ] 
face4 = [ corners[1], corners[2], corners[3] ] 
```

```python
v1 = np.array([face1,face2,face3,face4],dtype=float)
print(v1)
```

```python
objs.append(create_object(v1, [0.5,0.5,1]))
ax, fig = get_plots(objs)
```

## Rotating and Translation of 3D Objects

### Translation

```python
translation = np.array( [ 1, 0, 0 ], dtype=float ) 
v2 = v1 + translation
print(v2)
```

```python
objs.append(create_object(v2, [0.5,1,0.5]))
print(objs)
ax, fig = get_plots(objs)
```

### Rotation

```python
theta = np.pi/6
R = np.array( [ [ np.cos(theta), -np.sin(theta), 0 ],
                [ np.sin(theta),  np.cos(theta), 0 ],
                [ 0,              0,             1 ] ], dtype=float )
v3 = np.matmul(v1,R)
print(v3)
```

```python
objs.append(create_object(v3, [1,0.5,0.5]))
ax, fig = get_plots(objs)
```
